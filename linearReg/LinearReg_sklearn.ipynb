```python
# Linear Regression
# to analyse sale according to investment for TV ads
import numpy as np
data= np.array([[0,120],[50,195],[100,210],[150,275]])
```

```python
tv_ads = 200

from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(data[:,[0]],data[:,1])
print("the fitted line has coef",model.coef_)
print("the fitted line has slope",model.intercept_)

```

```python
sale=model.predict([[200]])
print("predicted sales",sale)
```
